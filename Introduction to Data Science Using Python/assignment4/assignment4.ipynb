```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import re


def nba_correlation():
    nba_df=pd.read_csv("assets/nba.csv")
    # extract just year 2018, remove rows containing Divison 
    nba_df=nba_df[nba_df.year==2018]
    
    # Extract last word of team
    pattern='(?P<team>[\w]*)[\*]*\s\(\d{1,2}\)$'
    def team_split(f):
        for item in re.finditer(pattern, f):
            return item.groupdict()['team']
    nba_df['team2']=nba_df['team'].apply(team_split)
    
    cities=pd.read_html("assets/wikipedia_data.html")[1]
    cities=cities.iloc[:-1,[0,3,5,6,7,8]]
    cities.rename(columns={'Population (2016 est.)[8]':'Population'},inplace=True)
    # reset indices
    cities.index=np.arange(len(cities.index))

    # strip out notes from cities NBA
    pattern='(?P<team>[\w]*)[\[note\s\d\]]*$'
    def team_split(f):
        for item in re.finditer(pattern, f):
            return item.groupdict()['team']
    cities=cities[['Metropolitan area','Population','NBA']]
    cities['NBA2']=cities['NBA'].apply(team_split)

    nba2_df=nba_df
    for i,x in enumerate(nba_df['team2']):
        if len(list(cities.index[cities['NBA2'].str.contains(x)==True]))>0:
            idx=int(list(cities.index[cities['NBA2'].str.contains(x)==True])[0])
            nba2_df.loc[i,'M']=cities['Metropolitan area'][idx]
            
    # Sort by Metro area
    cities.sort_values(by='Metropolitan area',inplace=True)
    cities2=cities[cities.NBA2!='']

    nba2_df.sort_values(by='M',inplace=True)
    nba2_df.W=nba2_df.W.astype(int)
    nba2_df.L=nba2_df.L.astype(int)
    
    # Group NHL data by Metro
    nba3_df=nba2_df.groupby(by=['M'])[['W','L']].sum()
    
    # Calculate Win/Loss Ratio fror NHL
    nba3_df['W'] = nba3_df['W'].astype(float)
    nba3_df['L'] = nba3_df['L'].astype(float)
    nba3_df['WL']= nba3_df['W'] + nba3_df['L']
    nba3_df['WL Ratio']= nba3_df['W']/nba3_df['WL']
    nba3_df['Metro area']=nba3_df.index
     
    # reorder index
    nba3_df=nba3_df.reindex(cities["Metropolitan area"])
    nba4_df=nba3_df[nba3_df['Metro area'].isnull()==False]

    #raise NotImplementedError()

    population_by_region = cities2['Population'].astype(float).to_list() # pass in metropolitan area population from cities
    win_loss_by_region = nba4_df['WL Ratio'].to_list() # pass in win/loss ratio from nhl_df in the same order as cities["Metropolitan area"]
    
    assert len(population_by_region) == len(win_loss_by_region), "Q2: Your lists must be the same length"
    assert len(population_by_region) == 28, "Q2: There should be 28 teams being analysed for NBA"

    return stats.pearsonr(population_by_region, win_loss_by_region)[0]
```
